```python
import serial
import time

port = "COM3"   # change this
baud = 921600   # Newport USB communication setting

ser = serial.Serial(
    port=port,
    baudrate=baud,
    bytesize=serial.EIGHTBITS,
    parity=serial.PARITY_NONE,
    stopbits=serial.STOPBITS_ONE,
    timeout=1,
    write_timeout=1,
)

def send(cmd, wait=0.05):
    """
    Send one Newport Agilis ASCII command.
    Commands must end with CR/LF.
    """
    ser.write((cmd + "\r\n").encode())
    time.sleep(wait)
    return ser.read_all().decode(errors="ignore").strip()

print("Firmware:", send("VE"))

# Put controller in remote mode
send("MR")

# Check errors
print("Error:", send("TE"))

ser.close()
```
